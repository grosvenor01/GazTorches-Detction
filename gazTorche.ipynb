### startthe machine learning model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset.csv")
df.columns = ['Feature1', 'Feature2','Gaz_Volume']
y = df["Gaz_Volume"]
df= df.drop("Gaz_Volume" , axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.05, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

### Image Processing here

```python
import cv2
import pandas as pd

image = cv2.imread('/content/Capture.PNG')
boxes = []
for i in range(len(df)):
    boxes.append([df["xmin"].iloc[i], df["ymin"].iloc[i], df["xmax"].iloc[i], df["ymax"].iloc[i], df["class"].iloc[i], df["predicted_volume"].iloc[i]])
print(len(df))
font = cv2.FONT_HERSHEY_SIMPLEX
for box in boxes:
    x1, y1, x2, y2, label, percentage = box
    if label == 1:
      cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (139, 69, 19), 2)
      cv2.putText(image, "gaz-torch", (int(x1), int(y1-10)), font, 0.6, (139, 69, 19), 2, cv2.LINE_AA)
      cv2.putText(image, '',(int(x1), int(y2+20)), font, 0.6, (139, 69, 19), 2, cv2.LINE_AA)
    else :
      cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
      percentage_text = f"{percentage*100:.2f}Litres"
      cv2.putText(image, percentage_text, (int(x1), int(y2+20)), font, 0.6, (0, 0, 255), 2, cv2.LINE_AA)

cv2.imwrite('labeled_image.png', image)
cv2.destroyAllWindows()
```

### Video Processing

```python
cap = cv2.VideoCapture('/content/video.mp4')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, fps, (width, height))


while True:
    ret, frame = cap.read() #treturni booolean + el frame as numpy array
    if not ret:
        break
    results = model(frame)
    df = results.pandas().xyxy[0]
    df['width'] = df['xmax'] - df['xmin']
    df['height'] = df['ymax'] - df['ymin']
    df['predicted_volume'] = 0
    for _, row in df.iterrows():
      x = np.array([row['height'], row['width']])
      x = x.reshape(1, -1)
      predicted_vol = ext_model.predict(x)
      df.at[_, 'predicted_volume'] = predicted_vol[0]
    boxes = []
    for i in range(len(df)): #same processing ta3 l'image (we do devid it to frames)
        boxes.append([df["xmin"].iloc[i], df["ymin"].iloc[i], df["xmax"].iloc[i], df["ymax"].iloc[i], df["class"].iloc[i], df["predicted_volume"].iloc[i]])

    font = cv2.FONT_HERSHEY_SIMPLEX
    for box in boxes:
        x1, y1, x2, y2, label, percentage = box
        if label == 1:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (139, 69, 19), 2)
            cv2.putText(frame, "gaz-torch", (int(x1), int(y1-10)), font, 0.6, (139, 69, 19), 2, cv2.LINE_AA)
            cv2.putText(frame, '',(int(x1), int(y2+20)), font, 0.6, (139, 69, 19), 2, cv2.LINE_AA)
        else:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
            percentage_text = f"{percentage*100:.2f}Litres"
            cv2.putText(frame, percentage_text, (int(x1), int(y2+20)), font, 0.6, (0, 0, 255), 2, cv2.LINE_AA)
    out.write(frame)

cap.release()
out.release()
```
